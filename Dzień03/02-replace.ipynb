```python
import pandas as pd
import numpy as np
```

```python
url='http://51.91.120.89/extras/travel.csv'
df = pd.read_csv(url)
df
```

```python
df.dtypes #sprawdzam typy danych
```

```python
#zamiana wartości 2000 i 5000 na inne wartosci
df.replace(['2000','5000'],np.NaN) 

# w tym przykladzie widzimy ze mozemy robic w replace wyliczenia. Dzieki temu nie musimy instrukcji stosowac kilkarazy
# nie podano inplace=true ale jest to mozliwe
```

```python
df.replace({"Age":["2000","5000"], 
           "travel_id": 1}, np.NaN) 

#df.replace({"nazwa_kolumny":["lista wartosci ktore nas interesuja"]}, np.NaN) 
#tutaj przyklad zamiany na kilku kolumnach na podstawie wartosci przekazanych w słowniku. Wartosci sa zamieniane na NaN 
#przy użyciu metody np.NaN
```

```python
df.replace({"2000":'----',
           '5000':'****'})

#łańcuchy znakow ktore wygladaja jak 2000 będą zamieniane na ---- a wartość 5000 na **** operacja zostanie zrealizowana 
#na całym df
# w tej procedurze można użyć map jednak będzie to powodowało w przypadku braku odniesienia w podanym słowniku 
# -> będzie wprowadzone NaN dla brakujących odniesień 
```

```python
#teraz przykład z regular expressions
df
```

```python
#Teraz przyklad jak odnalezc wzorzec literek i go usunac z danych
# jak z regular expression wykonac zadanie - wyrazenie re zbierajace a-z i spacje:
# [A-Za-z ]
df.replace({"No_of_pkg" : '[A-Za-z ]'},"",regex=True) #najpierw podaje gdzie ma szukac potem z regex wskazuje czego ma szukać (pattern wg. regex) 
```

```python
# jak przy użyciu jednek konstrukcji replace usunac kilka par obiektow stare wartosci na nowe wartosci
# przy uzyciu replace mozna zamienic oceny slowne na liczbowe
data = {
    "imie": ['Jan','Maria','Rokita'],
    "ocena":['dst','bd','dobry']
}
df=pd.DataFrame(data)
df
```

```python
df.replace(['dst','bd','dobry'],[3,5,4]) #odbywa sie zamiana wazne aby elementy pierwszej tablicy mialy przypisane elemeny
#w drugiej tablicy
```
