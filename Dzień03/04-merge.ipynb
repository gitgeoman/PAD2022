```python
import pandas as pd
```

```python
url1 = 'http://51.91.120.89/extras/otodom-head-sample.csv'
df1 = pd.read_csv(url1, sep=';')
df1=df1[df1.columns[3:]] #wybieram kilka kolumn
df1
```

```python
url2='http://51.91.120.89/extras/otodom-details-sample.csv'
df2 = pd.read_csv(url2, sep=';', usecols=['ID','descr'])
df2.head()
```

```python
#w celu polączeni musze mieć klucz wspólny -> będzie nim ID
```

```python
df1.shape, df2.shape
```

```python
#złączenie typu inner na podstawie kolumny id
pd.merge(df1, df2, on="ID")
```

```python
#zmiana nazwy kolumny z ID na ogl_ID
```

```python
#złączenie typu inner na podstawie kolumny on, left_on, right_on -> kolumny która ma 
pd.merge(df1, df2, on="ID")
```

```python
#zmiana nazwy kolumny z ID na ogl_ID żeby spreparować sytuacje gdy nazwy kolumn sobie nie 
#odpowiadają mimo że mają te same daen
df2.rename(columns={"ID":"ogl_ID"},inplace=True)
```

```python
#złączenie typu inner na podstawie kolumny id, of=ogl_id
pd.merge(df1, df2, left_on="ID", right_on="ogl_ID", how='inner')
```

```python
#złączenie na lewą strone na podstawie wspólnych danych
df_res = pd.merge(df1, df2, left_on="ID", right_on="ogl_ID", how='left') 
#wychodzi 300 wierszy
df_res
```

```python
df_res[df_res.descr.isna()] #wyciagam trzy wiersze ktore nie maja odniesienia w drugim zbiorze
```

```python
#złączenie na lewą strone na podstawie wspólnych danych
df_res = pd.merge(df1, df2, left_on="ID", right_on="ogl_ID", how='left', indicator=True) 
#indicator dodaje nowa kolumne w ktorej zapisuej informacje o tym jak wygladało złączenie
df_res
```

```python
df_res._merge.value_counts() # opis tego jak wyglądało połączenie (kolumna ma nazwe _merge)
```

```python
df_res.query('_merge=="left_only" ') # opis innego rodzaju
```

```python
#złączenie na prawa strone na podstawie wspólnych danych
df_res = pd.merge(df1, df2, left_on="ID", right_on="ogl_ID", how='right', indicator=True) 
df_res
```

```python
df_res._merge.value_counts() # opis tego jak wyglądało połączenie (kolumna ma nazwe _merge)
#wszystkie złąza powstały w wyniku łączenia - nie ma rekordów które "wiszą"
```

### outer join na podstawie generowanych danych ćwiczenie praktyczne

```python
#metoda sample - umożliwa wziecie losowej ilosci danych z naszego df 
```

```python
#losuje dane z obu repozytoriów (33%) ustawiam random state żeby mieć te same wyniki
df1 = df1.sample(frac=.3, random_state=0)
df2 = df2.sample(frac=.3, random_state=0)
df1.shape, df2.shape
```

```python
#złączenie obustornne strone na podstawie wspólnych danych
df_res = pd.merge(df1, df2, left_on="ID", right_on="ogl_ID", how='outer', indicator=True)
#indicator dodaje kolumne opisujaca sposob polaczenia
df_res
```

```python
df_res._merge.value_counts()
```
