### Duplikaty

```python
import pandas as pd
```

```python
#klucze to nazwy kolumn w df wartosci to wartosci, df mozna utworzyc z listy, elementami listy moga być obiekty
data = [{"imie":"Jan", "nazwisko":"Kowalski","ocena":4.76, "klasa":4},
        {"imie":"Andrzej", "nazwisko":"Nowak","ocena":4.0, "klasa":5},
        {"imie":"Adam", "nazwisko":"Zieliński","ocena":5.0, "klasa":6},
       {"imie":"Jan", "nazwisko":"Kowalski","ocena":4.76, "klasa":4},
       {"imie":"Andrzej", "nazwisko":"Nowak","ocena":4.0, "klasa":5},
       {"imie":"Jan", "nazwisko":"Pawlak","ocena":5.0, "klasa":6,"zachowanie":'dobre'}]

df = pd.DataFrame(data) #<- struktura zewnetrzna 
df
```

    df.duplicated() <- służy do sprawdzenia czy są duplikaty, zwracany jest obiekt serii, na tym obiekcie trzeba zrobić sumowanie i mamy informacje ile tycho obiektow jest, aby dowiedziec sie na ktorej pozycji wystepuja uzywamyy innej metody

```python
df.duplicated().sum() # okreslenie ilosci duplikatow
```

```python
#pokaz zduplikowane rekordy
df[df.duplicated()]
#ciekawa sztuczka ponieważ df.duplicated() zwraca serie w ktorej rekordy to prawda lub falsz i w wyniku podczytania takiej listy dostajemy duplikaty
```

```python
#usuwanie duplikatów
df.drop_duplicates() #shift tab tab otwiera dukumentacje
```

```python
#usuwanie duplikatów
df.drop_duplicates(keep='first', ignore_index=True) #keep zostawi pierwszy obiekt ktory sie powtarza, ignore)index - od nowa po usunieciu rekordordow
```

```python
#usuwanie duplikatów
df.drop_duplicates(keep='last' ) #keep zostawi pierwszy obiekt ktory sie powtarza,
#powyższe przyda się gdy chcemy ostatni obiekt z listy danych
```

```python
df.drop_duplicates(keep=False, ) #keep =False usuwa wszystkie dane zostawia duplikaty.
```

```python
df.drop_duplicates(subset='imie', keep='first') #subset umożliwia wyłapytanie duplikatów na podstawie wartości w kolumnach
```

### wypełnianie wartości nan

```python
url = 'http://51.91.120.89/extras/season.csv'
df = pd.read_csv(url)
df
```

```python
df.dtypes
```

jeżeli w danych jest nan to liczbami będzie automatycznie float 64

```python
#wypelnianie wartości metody FILL NA
df.fillna(0) #metoda działa na calym df
#ale mozna jej tez uzyc na pojedynczej serii danych
df.day.fillna("---") #tutaj żeby się odrazu nadpisało trzeba dać inplace true w innym przypadku dostajemy kopie serii danych
```

```python
#teraz wartosc nan zostanie zastapiona przez rekord ktory poprzedza
df.fillna(method='ffill')
```

```python
#teraz wartosc nan zostanie zastapiona przez rekord ktory poprzedza, limit propagacji można ograniczyć przy użyciu limit
df.fillna(method='ffill', limit =1) # w ten sposób tylko o 1 zostanie wypropagowane
```

```python
#teraz wartosc nan zostanie zastapiona przez rekord ktory następuje po, (propagacja wsteczna)
df.fillna(method='bfill', ) 
```

```python
#jest możliwa interpolacja na podstawie wartosci poprzedzajacych i następujacych
```

### czynnosci do zrobienia przy oczyszczanu na
 <li>w kolumnie "day zmieniamy NAN na znak ---
 <li>w kolumnie "temp" zamiana NAN na wartosc srednia
 <li>w kolumnie "windspeed" not available na NAN a NaN na średnią 
     <li>w kolumnie dates konwersja na typ datatime64

```python
df.day.fillna('---', inplace=True)
df
```

```python
#w kolumnie "temp" zamiana NAN na wartosc srednia
df.temp.mean() #>obliczy wartosc srednia - metoda omieja nan przy obliczeniach
```

```python
df.temp.fillna(df.temp.mean(), inplace=True)
df
```

```python
#w kolumnie "windspeed" not available na NAN a NaN na średnią
#1. zmieniam konwertuje not available na NAN
#2 zeby obliczyc srednio musze zrobic konwersje kolumny widspeed bo typ tej kolumny to object -> do typu numerycznego
#3. obliczam srednia i zastepuje nan srednia
#4. konwertuje kolumne dates do typu data 


df['wind-speed'].replace('Not available', pd.np.NaN, inplace=True) #1
df['wind-speed'] = df['wind-speed'].astype('float32') #2
df['wind-speed'].fillna(df['wind-speed'].mean(), inplace=True) #3
df.dates = pd.to_datetime(df.dates,format='%m/%d/%Y')
```
