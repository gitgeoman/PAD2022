### łączenie danych z kilku df
### metoda append
- uproszczone działanie względem concat
### metoda concat 
- może działać na wierszach i na kolumnach, concat umożliwia dokonanie operacji na więcej niz na 2 df. 

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'a':[1,2,3,4,],"b":[5,6,7,8],"c":[9,10,11,12],"d":[13,14,15,16],'e':[17,18,19,20],})
df1
```

```python
df2 = pd.DataFrame({'a':[1,2,3,4,],"b":[5,6,7,8],"c":[9,10,11,12],"d":[13,14,15,16],"f":[17,18,19,20],})
df2*=2
df2
```

```python
df1.append(df2, ignore_index=True) #metoda append dodaje wiersze do kolejenej 
# w wyniku pracy tworzony jest nowy obiekt - w metodzie nie ma paramteru inplace
# w metodzie append gdy trzeba ją wywołać kilkaset razy to trzeba dodać ignore index co przyśpieszy konkatenacje
#jezeli dane nie majać identycznych nazw kolumn brakujace dane są uzupełniane NaN, przy czym następuje konwersja 
#do innego typu obiektów
```

### Metoda Concat

```python
df1 = pd.DataFrame({
    "A":["A"+str(x) for x in range(4)],
    "B":["B"+str(x) for x in range(4)],
    "C":["C"+str(x) for x in range(4)],
    "D":["D"+str(x) for x in range(4)]
})
df1
```

```python
df2 = pd.DataFrame({
    "A":["A"+str(x) for x in range(4,8)],
    "B":["B"+str(x) for x in range(4,8)],
    "C":["C"+str(x) for x in range(4,8)],
    "D":["D"+str(x) for x in range(4,8)]
})
df2
```

```python
pd.concat([df1,df2],axis=0, ignore_index=True) #łącze dwie df pionowo od nowa zliczam index
```

```python
pd.concat([df1,df2],axis=1,)
```

```python
df1 = pd.DataFrame({
    "A":["A"+str(x) for x in range(4)],
    "B":["B"+str(x) for x in range(4)],
    "C":["C"+str(x) for x in range(4)],
    "D":["D"+str(x) for x in range(4)]
})
df2 = pd.DataFrame({
    "A":["A"+str(x) for x in range(4,8)],
    "B":["B"+str(x) for x in range(4,8)],
    "C":["C"+str(x) for x in range(4,8)],
    "D":["D"+str(x) for x in range(4,8)]
}, index=[3,4,5,6])
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1) 
#tutaj "centrowanie" (dołącznie) danych jest prowadzone na podstawie 
#indeksow roznica w zbiorach danych został uzupełniona NaN
```

```python
pd.concat([df1,df2],axis=1, join='inner') #<-tylko gdy dane mają pełne pokrycie (czesc wspolna dwoch zbiorow)
```

```python
pd.concat([df1,df2],axis=1, join='outer') #<-tylko gdy dane mają pełne pokrycie
```
