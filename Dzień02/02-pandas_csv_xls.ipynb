```python
#wymuszanie typu dla wywoÅ‚ywanych kolumn #optymalizacja
#import danych z sieci obiektu df 
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url, dtype={
    'rank': "uint16",
    'revenue_change' : 'float32'
}) # wczytywanie danych z csv
df.dtypes
```

```python
df['rank'].min(), df['rank'].max()
```

```python
#co siÄ™ dzieje gdy przesadziliÅ›my z optymalizacjÄ… - nic - zostanie ona przeprowadzona ale zostanie 
#odciÄ™cie wartoÅ›ci powyÅ¼ej zakresu zmiennej
#wymuszanie typu dla wywoÅ‚ywanych kolumn #optymalizacja

#import danych z sieci obiektu df 
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url, dtype={
    'rank': "uint8", #tutaj jest przesadzone z 'downsizingiem'
    'revenue_change' : 'float32'
}) # wczytywanie danych z csv
df.dtypes
```

```python
#
df['rank'].min(), df['rank'].max()
```

```python
#konkluzja jest taka Å¼e trzeba sprawdzaÄ‡ wartoÅ›ci minimalne i maksymalne (scope wartoÅ›ci) danych wejÅ›ciowych.
#taka oepracja jest moÅ¼liwa dla caÅ‚ego zbioru danych przy uÅ¼yciu describe
#ciekawe Å¼e sÄ… tu podane kwantyle zbiorÃ³w danych
# count pokazuje ile danych ma wypisane wartoÅ›ci (np. rank jest wszystko wypeÅ‚nione a w profit_change - ma missing value NaN)
df.describe()
```

```python
#podczas importu danych jest moÅ¼liwa implementacja konwerterÃ³w (funkcji ktÃ³re przetworzÄ… dane wejciowe)

#konwersja w momencie tworzenia
#krok 1 - definiuje funkcje do konwersji ktÃ³re bÄ™dÄ™ wywoÅ‚ywaÅ‚
#krok
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url, converters={
    #podaje nazwe kolumny na ktÃ³rej ma dziaÅ‚aÄ‡ funkcja do konwersji
    "rank": lambda x: int(x)<=10
}) 
df.head(5)
# to jest metoda ktÃ³rÄ… warto robiÄ‡ przy gdy mamy maÅ‚e repozytorium.
#gdy sa duÅ¼e repozytoria to lepiej wczytaÄ‡ dane do pamiÄ™ci i zajÄ…Ä‡ siÄ™ konwersjÄ… przy uÅ¼yciu apply.
# jest tak dlatego Å¼e pandas przetwarza wszystko w pamiÄ™ci. To dziÄ™ki temu pandas jest taki szybki. 
#Gdy pandas ma za duÅ¼o danych to trzeba swipowaÄ‡ dane na dysk twardy (przerzucaÄ‡ dane)
# wtedy trzeba zwrÃ³ciÄ‡ uwagÄ™ Å¼e spada wydajnoÅ›Ä‡ w wyniku takiej operacji
#
#
#
#biblioteki ktÃ³re radzÄ… sobie z duÅ¼ym zbiorem danych to moÅ¼emy uÅ¼yÄ‡ daska lub sparka
```

```python
#ustawianie wÅ‚asnego indeksu na podstawie danych z kolumny df na przykÅ‚adzie wybranej kolumny 
#moÅ¼na to zrobiÄ‡ w momencie przekazywania danych do df
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url,index_col='company') 
df.head(10)
# trzeba zwrÃ³ciÄ‡ uwagÄ™ Å¼e to nie jest juÅ¼ kolumna tylko index
```
