```python
#wymuszanie typu dla wywoływanych kolumn #optymalizacja
#import danych z sieci obiektu df 
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url, dtype={
    'rank': "uint16",
    'revenue_change' : 'float32'
}) # wczytywanie danych z csv
df.dtypes
```

```python
df['rank'].min(), df['rank'].max()
```

```python
#co się dzieje gdy przesadziliśmy z optymalizacją - nic - zostanie ona przeprowadzona ale zostanie 
#odcięcie wartości powyżej zakresu zmiennej
#wymuszanie typu dla wywoływanych kolumn #optymalizacja

#import danych z sieci obiektu df 
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url, dtype={
    'rank': "uint8", #tutaj jest przesadzone z 'downsizingiem'
    'revenue_change' : 'float32'
}) # wczytywanie danych z csv
df.dtypes
```

```python
#
df['rank'].min(), df['rank'].max()
```

```python
#konkluzja jest taka że trzeba sprawdzać wartości minimalne i maksymalne (scope wartości) danych wejściowych.
#taka oepracja jest możliwa dla całego zbioru danych przy użyciu describe
#ciekawe że są tu podane kwantyle zbiorów danych
# count pokazuje ile danych ma wypisane wartości (np. rank jest wszystko wypełnione a w profit_change - ma missing value NaN)
df.describe()
```

```python
#podczas importu danych jest możliwa implementacja konwerterów (funkcji które przetworzą dane wejciowe)

#konwersja w momencie tworzenia
#krok 1 - definiuje funkcje do konwersji które będę wywoływał
#krok
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url, converters={
    #podaje nazwe kolumny na której ma działać funkcja do konwersji
    "rank": lambda x: int(x)<=10
}) 
df.head(5)
# to jest metoda którą warto robić przy gdy mamy małe repozytorium.
#gdy sa duże repozytoria to lepiej wczytać dane do pamięci i zająć się konwersją przy użyciu apply.
# jest tak dlatego że pandas przetwarza wszystko w pamięci. To dzięki temu pandas jest taki szybki. 
#Gdy pandas ma za dużo danych to trzeba swipować dane na dysk twardy (przerzucać dane)
# wtedy trzeba zwrócić uwagę że spada wydajność w wyniku takiej operacji
#
#
#
#biblioteki które radzą sobie z dużym zbiorem danych to możemy użyć daska lub sparka
```

```python
#ustawianie własnego indeksu na podstawie danych z kolumny df na przykładzie wybranej kolumny 
#można to zrobić w momencie przekazywania danych do df
url ='http://51.91.120.89/extras/f500.csv'
df =  pd.read_csv(url,index_col='company') 
df.head(10)
# trzeba zwrócić uwagę że to nie jest już kolumna tylko index
```
