```python
import pandas as pd
import numpy as np
```

```python
url = 'http://51.91.120.89/extras/nba.csv'
df = pd.read_csv(url, index_col='Name') # wczytuje dane ustawiam jako index name zawodnika
df
```

```python
#Oczyszczam dane -> szukamy NaN
df.isna().sum()
```

```python
#z powyÅ¼szego pliku tekstowego CSV bÅ‚Ä™dnie wczytaÅ‚a siÄ™ jedna linijka (wiersz) 
#-> wszystkie jej pola zostaÅ‚a zinsterpretowana jako Nan - trzeba siÄ™ tego pozbyÄ‡

#oczyszczanie zbioru danych z NaN - poniÅ¼ej usuwa wiersze z wartoÅ›ciami Nan
df.dropna(axis=0, how='all', inplace=True) #inplace nadpisuje df (bez koniecznoÅ›ci pisania df=df.dropna.....)
df
```

```python
#szukamy NaN w wynikach 
df.isna().sum() #metoda do sprawdzenia gdzie jest nan +#metoda do sumowania
```

```python
#zamiana Nan w college na string pusty
df["College"].replace(np.NaN,'', inplace=True) #replace(co zastÄ…piÄ‡, czym, nadpisywanie)
df.head(5)
```

```python
#konwersja kolumn wagi 
df['Weight']/=2.2
```

```python
#konwersja wysokoÅ›ci
def convert_height(text):
    feet,inches= text.split('-')
    return int(feet)*30.4+int(inches)*2.54

df.Height.apply(convert_height) #poniewaÅ¼ zamiana jest bardziej zÅ‚oÅ¼ono uÅ¼yto metody apply
df.head(5)
```

```python
#zapis wynikÃ³w do csv
df.to_csv("nba1.csv")
#zapis wynikÃ³w do xlsx
df.to_excel('nba.xlsx')
```
