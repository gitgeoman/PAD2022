```python
import pandas as pd
import numpy as np
```

```python
url = 'http://51.91.120.89/extras/nba.csv'
df = pd.read_csv(url, index_col='Name') # wczytuje dane ustawiam jako index name zawodnika
df
```

```python
#Oczyszczam dane -> szukamy NaN
df.isna().sum()
```

```python
#z powyższego pliku tekstowego CSV błędnie wczytała się jedna linijka (wiersz) 
#-> wszystkie jej pola została zinsterpretowana jako Nan - trzeba się tego pozbyć

#oczyszczanie zbioru danych z NaN - poniżej usuwa wiersze z wartościami Nan
df.dropna(axis=0, how='all', inplace=True) #inplace nadpisuje df (bez konieczności pisania df=df.dropna.....)
df
```

```python
#szukamy NaN w wynikach 
df.isna().sum() #metoda do sprawdzenia gdzie jest nan +#metoda do sumowania
```

```python
#zamiana Nan w college na string pusty
df["College"].replace(np.NaN,'', inplace=True) #replace(co zastąpić, czym, nadpisywanie)
df.head(5)
```

```python
#konwersja kolumn wagi 
df['Weight']/=2.2
```

```python
#konwersja wysokości
def convert_height(text):
    feet,inches= text.split('-')
    return int(feet)*30.4+int(inches)*2.54

df.Height.apply(convert_height) #ponieważ zamiana jest bardziej złożono użyto metody apply
df.head(5)
```

```python
#zapis wyników do csv
df.to_csv("nba1.csv")
#zapis wyników do xlsx
df.to_excel('nba.xlsx')
```
