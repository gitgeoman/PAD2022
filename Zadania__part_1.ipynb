 która w minimalnej liczbie kroków tworzy (i zwraca) kwadratową tablicę
NumPy zawierającą na przemian jedynki i zera - w postaci jak na grafice -
 o dowolnym  rozmiarze (zadanym przez argument wywołania),
Powodzenia!

```python
import numpy as np
##ROZWIĄZANIE list comprehension z if statment 
def jeden_zero(rozmiar):
  return np.array([[1 if x%2==0 else 0 for x in range(rozmiar)] for i in range(rozmiar)])

jeden_zero(4),  jeden_zero(6), jeden_zero(10)
```

   c) 'workers' rosnąco

```python
import pandas as pd

def intToRoman(num):
  
    # Storing roman values of digits from 0-9
    # when placed at different places
    m = ["", "M", "MM", "MMM"]
    c = ["", "C", "CC", "CCC", "CD", "D",
         "DC", "DCC", "DCCC", "CM "]
    x = ["", "X", "XX", "XXX", "XL", "L",
         "LX", "LXX", "LXXX", "XC"]
    i = ["", "I", "II", "III", "IV", "V",
         "VI", "VII", "VIII", "IX"]
  
    # Converting to roman
    thousands = m[num // 1000]
    hundreds = c[(num % 1000) // 100]
    tens = x[(num % 100) // 10]
    ones = i[num % 10]
  
    ans = (thousands + hundreds +
           tens + ones)
  
    return ans

df = pd.read_csv("https://www.dropbox.com/s/axeucvys209d30h/inc5000_all10years.csv?dl=1").assign(year_roman = lambda x: x.year)
df.year_roman = df.year_roman.apply(intToRoman)
df
```

```python
#Filtrowanie i Sortowanie
df[((df.industry=='Software')|(df.state_s=='CA'))&(df.year==2017)].sort_values(['yrs_on_list','revenue','workers'], ascending=[True,False,True])
```
