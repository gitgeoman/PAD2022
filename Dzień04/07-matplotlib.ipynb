```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#https://color.adobe.com/pl/create/color-wheel <-wybór kolorków
x = np.arange(0,4.5*np.pi,.2)
x
```

```python
y = np.sin(x)
y
```

```python
#rysuje wykres
plt.figure(figsize=(15,6)) # zmiana wielkości wykresu
plt.plot(
    x,#oś x
    y,#oś y
    color='red', #kolor linii
    linestyle='--', #styl linia przerywana
    marker ='s', #znacznik
    markersize=6 #wielkosc znacznika
)
plt.grid(); # dodaje siatke
plt.xlabel('x'); # etykietka dla osi poziomej
plt.ylabel('sin(x)');# etykietka dla osi pionowej
plt.title('Wykres sin(x)')
plt.xticks(#do określenia wartości które maja byc na osi x
    np.arange(0,4.5*np.pi, np.pi), #generuje przy użyciu arange wartości które mają być wyświetlane
    ['0','π','2π','3π','4π'] # etykietki dla wartości
);
```

```python
#Rysowanie wykresu punktowego 
#(1,1) (2,5) (3,8) (4,10) <-to są współrzędne punktów
plt.plot(
    [1,2,3,4], # x współrzędnych
    [1,5,8,10], #y współrzędnych
    'ro' # r - red o - oval 
)
plt.axis([0,5,0,15]) #zakres wartosci na osiach od 0-5 dla x; 0-15 dla y
plt.show()
```

```python
#zadanie 
#narysuj kilka wykresow na jednym rysunku dodaj strzałkę do wybranego punktu wpisz adnotacje i tekst na wykresie
#podobnie jak poprzednio żeby nanieść kilka osi danych 

#wykres 1
x = np.arange(0,5.2,.2)
plt.figure(figsize=(10,6))
plt.plot(x,x,'g--')
#wykres 2
plt.plot(x, 0.5*x**2, 'k-.')
#wykres 3
plt.plot(x, 1/8*x**3,'r')
plt.text(0,12,'tekst na wykresie', fontsize=30,color='gray') #koordynaty gdzie ma być napis, co ma być napisane

#strzałka z adnostacją 
plt.annotate(
    'adnotacja na wykresie', 
    [2,2], 
    [3,10], 
    arrowprops={ # stylizacja strzałki
        'linewidth':1, #szerokość linii
        'arrowstyle':'<-' #kierunek grotu jest ważny mozna też '<->', '->'
    }
)
plt.show()
```

```python
#powyższy zapis można uprościć i zapisać w jednej linijce
plt.plot(x,x,'g--',x, 1/8*x**3,'r',x, 0.5*x**2, 'k-.');
```

### wykres słupkowy

```python
countries = ['USA','Rosja', 'Kanada'] #listy
m_bronze = np.array([12,10,8]) #tablice numpy
m_silver = np.array([4,2,8])
m_gold = np.array([2,2,1])
```

```python
# plt.figure(figsize=(10,6)) # zmieniam rozmiar wykresu poniżej alternatywny sposób !!!
plt.bar(
    countries, #oś x
    m_bronze,  #oś y
    width=.5, #przerwy między paskami + szerokosc
    color='brown',
    label='Bronze'
)

plt.bar(
    countries, #oś x
    m_silver,  #oś y
    width=.5, #przerwy między paskami + szerokosc
    color='silver',
    bottom=m_bronze # atrybut aby słupki znalazły się nad sobą a nie wszystkie były od tego samego poziomu '0'
    #z tego powodu m_bronze musi być tablica numpy
    ,label='Silver'
)

plt.bar(
    countries, #oś x
    m_gold,  #oś y
    width=.5, #przerwy między paskami + szerokosc
    color='gold',
    bottom=m_bronze+m_silver, # atrybut konieczny, aby słupki znalazły się nad sobą a nie wszystkie były od tego samego poziomu '0'
    #ponieważ tu jestdodawanie wartosci z m_bronze i m_silver to dlatego na początku trzeba było zrobić tablice numpy
    label='Gold'
)

#poniżej stylizacja legendy
plt.legend(
    loc= 'upper right', #metoda steruje miejscem gdzie jest legenda domyslnie jest to zlokalizowane tak aby niczego nie zasłaniało
                        #jeżeli chcialibyśmy żeby była na górze po prawej to używamy specjalnego atrybutu loc= 'upper right'
    ncol=2 #ilosc kolumn w ilu ma być legenda 
) 

plt.title('Statystyka medalowa')

plt.gcf().set_size_inches(10,6) # !! alternatywny sposób zmiany rozmiaru okna wykresu ->get canvas figsize, set size i podaje krotke

plt.savefig(
    'wykres.jpg', # jaka nazwa, jaki format zapisu danych
    dpi=300, #rozdzielczosc
)

plt.show()
```

### Wykres zbity (zestackowany)

```python
#Krok 1
#tworzymy listy z danymi
countries = ['USA','Rosja', 'Kanada'] #lista 
m_bronze = np.array([12,10,8]) #tablice numpy w przypadku wykresu słupkowego (jeden bar nad drugim musi być np.array)
m_silver = np.array([4,2,8])#tablice numpy w przypadku wykresu słupkowego (jeden bar nad drugim musi być np.array)
m_gold = np.array([2,2,1])#tablice numpy w przypadku wykresu słupkowego (jeden bar nad drugim musi być np.array)

#krok 2
#123 to pola dla usa i medali 4 to przerwa 5,6,7 to medale rosja
x=np.array([1,5,9]) #to jest oś x

#krok 3 
#określam rozmiar wykresu
plt.figure(figsize=(10,6)) 

#krok 4
#generuje słupki wykresów (w klasycznym wykresie osie)

#generuje oś dla wykresu
plt.bar(
    x, # dane oś x
    m_bronze,  #oś y
    width=.5, #przerwy między paskami + szerokosc
    color='brown',
    label='Bronze'
)
#generuje oś dla wykresu
plt.bar(
    x+0.5, #dane oś x
    m_silver,  #oś y
    width=.5, #przerwy między paskami + szerokosc
    color='silver',
    label='Silver'
)
#generuje oś dla wykresu
plt.bar(
    x+1, #oś x
    m_gold,  #oś y
    width=.5, #przerwy między paskami + szerokosc
    color='gold',
    label='Gold'
)

#poniżej stylizacja legendy
plt.legend(
    loc= 'best', #metoda steruje miejscem gdzie jest legenda domyslnie jest to zlokalizowane tak aby niczego nie zasłaniało
                        #jeżeli chcialibyśmy żeby była na górze po prawej to używamy specjalnego atrybutu loc= 'upper right'
    ncol=2 #ilosc kolumn w ilu ma być legenda 
) 

plt.title('Statystyka medalowa')

#krok 2 
#użyć xticks do wskazania poprawnych wartości na osi x bo na chwile obecno pochodzi ona z x

plt.xticks(x+.5, countries)
plt.show()
```

### Wykres kołowy

```python
names = ['Jaś', 'Małgosia','Zuzia','Julka']
toys = [10,8,18,6]
colors= ['blue', 'yellowgreen','#ff0033','lightskyblue']

plt.figure(figsize=(9,7))

plt.pie(
    x=toys, #dane wejsciowe
    labels=names, #etykietki danych
    colors=colors, # podaje tablice dla kolorow
    shadow = .1, #cień ala efekt perspektywy 3d
    autopct= '%.0f', #tutaj trzeba uważać bo zaokrąglenie powoduje że nie zawsze zsumuje się do stu!!!!!!!!!!!!!!!
    explode=[0.2,0,0,0], #odciaga na odległość 0.2 promienia okregu
    
)

plt.show()
```

### Wykres rozrzutu wartości (scatterplot) i

```python
import numpy as np

#generuje dane
np.random.seed(0)
N=100
x=6*np.random.randn(N)-3
y=2*x+np.random.randn(N)*5
```

```python
#tworze wykres rozrzutu wartosci
#krok 1 ustalam wielkosc wykresu
plt.figure(figsize=(7,7))

#krok 2 generuje 
plt.scatter(
    x, 
    y,
    c='r', #czerwony
    marker='s' #kwadraty w punktach
)
```

 ### histogram danych

```python
#zadanie zorobić dwa wykresy
plt.hist(
    x,  #dane z osi x
    bins=10,
    color='green'
);

plt.hist(
    y,  #dane z osi y
    bins=10,
    color='blue',
    alpha=0.3
);
```
