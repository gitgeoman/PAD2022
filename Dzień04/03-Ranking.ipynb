```python
import pandas as pd
#biblioteka pandas zawiera wbudowane narzędzia do tworzenia rankingow. 
```

```python
url ='http://51.91.120.89/extras/rank.csv'
df = pd.read_csv(url, usecols=[1,2,3])
df
```

```python
df.Score.rank() #otrzymujemy serie ktora zawiera informacje o rankingu
df['ranking'] = df.Score.rank()  # wynik zapisuje w postaci nowej kolumny metoda dziala tak że najwyzsza pozycja rankingu jest przypisana do najmniejszej wartosci 
df['ranking'] = df.Score.rank(ascending=False)  # żeby było normalnie najwiecej punktow na 1 miejscu ->
#tutaj trzeba zwrocic uwage ze taki ranking w przpadku pozycji egezekwo przypisywane jest 2.5
#mozna to zmienic przy uzyciu parametru method w raknigu ktora domyslnie liczy srednią z wszystkich wartosci

df.sort_values('ranking')
```

```python
df['ranking'] = df.Score.rank(
    ascending=False, #najwiecej punktow = najwyższa pozycja rankingu
    method ='min'     #egzekwo będą mieli najmniejsze wartosci
) 
#gdy sa pozycje egzekwo to wszyscy ktorzy sa egzekwo maja wartosc najmniejsza z
df.sort_values('ranking')
```

```python
df['ranking'] = df.Score.rank(ascending=False, method ='max') 
#gdy sa pozycje egzekwo to wszyscy ktorzy sa egzekwo maja wartosc najwieksza z miejsc ktore zajieliby zawodnicy
df.sort_values('ranking')
```

```python
df['ranking'] = df.Score.rank(ascending=False, method ='dense') 
#gdy sa pozycje egzekwo to wszyscy ktorzy sa egzekwo zajmuja miejsce na ktorym wypadli a po nim nie ma pominiec w rankingu
df.sort_values('ranking')
```

```python
df['ranking'] = df.Score.rank(ascending=False, method ='first') 
#gdy sa pozycje egzekwo to wszyscy ktorzy sa egzekwo uzyskuja lokate na podstawie wyniku a potem na podstawie wystapienia
df.sort_values('ranking')
```
