```python
import pandas as pd
import numpy as np
```

### Dyskretyzacja

```python
#generuje dane 
np.random.seed(0) #ustawiamy ziarno pseudolosowosci aby były te same dane
age = np.random.randint(0,106, 40)# tworze tablice do ktorej dodam dane losow wygenerowane z przedziału 0, 105 a probek ma byc 40
age
```

```python
#przeksztalcam tablice na df
df = pd.DataFrame(
    age, #podaje dane do przeksztalcenia
    columns=['Wiek'] #nadaje nazwe kolumn
)
df
```

```python
#dyskretyzacja opiera sie na wywolaniu metody cut na wybranej kolumnie
pd.cut(
    x=df.Wiek, # dane wejsciowe to kolumna
    bins=[0,18,65,110] # to są te nasze "koszyczki" (grupy)
)
#domyslnie przedzialy so lewostronnie otwarty i prawostronnie domkniety
#można to zmienic przy uzyciu parametru rigth
pd.cut(
    x=df.Wiek, # dane wejsciowe to kolumna
    bins=[0,18,65,110], # to są te nasze "koszyczki" (grupy)
    right=False #odwrocenie sposobu zamykania nawiasow
)
#można dodać etykiety przy użyciu atrybutu labels -> dzieki temu zdyskretyzowalismy dane
pd.cut(
    x=df.Wiek, # dane wejsciowe to kolumna
    bins=[0,18,65,110], # to są te nasze "koszyczki" (grupy)
    right=False, #odwrocenie sposobu zamykania nawiasow
    labels=['Dzieci','Dorosli','Seniorzy']
)
```

```python
#przypisuje wartosc operacji do nowej kolumny
df['zakres_wiekowy']=pd.cut(
    x=df.Wiek, # dane wejsciowe to kolumna
    bins=[0,18,65,110], # to są te nasze "koszyczki" (grupy)
    right=False, #odwrocenie sposobu zamykania nawiasow
    labels=['Dzieci','Dorosli','Seniorzy'] # podaje etykietki
)
df.head(10)
```

## One-hot encoding - jedno "gorące" a reszta "zimne"
- czesto stosowane w uczeniu maszynowym. w uczeniu maszynowym trzeba podawac liczby.

```python
np.random.seed(0)
data = np.random.choice(['Pn','Wt','Śr'], 20) #generuje tablice z podanych wartosci
df = pd.DataFrame(data, columns=['Dzien_tyg'])
df
```

```python
#robimy macierz indykatorow binarnych 
#->w tym celu tworzymy trzy dodatkowe kolumny ktore beda nazywaly sie pn, wt, sr i gdy bede mial pn to w kolumnie bedzie 
# chodzi o to ze jak zamienimy pn, wt, sr na cyferki to algorytm uczenia maszynowego bedzie szukal miedzy nimi zalezności.
```

```python
pd.get_dummies(df.Dzien_tyg) # generuje tablice indykatorow binarnych na podstawie wartosci dyskretnych
```

```python
pd.concat(
    [df,pd.get_dummies(df.Dzien_tyg)],
    axis=1
) #łącze dane żeby lepiej wyglądało -> to jest właśnie one hot encoding 
#czyli gdy chcemy zakodować dane dyskretne w postaci indykatorow
```
